```python
import random

class Numbers:
    def __init__(self):
        pass

    def Play(self):
        chosenNum = random.randint(1,20)

        for i in range(3):
            guess = input("Enter Guess: ")
            if chosenNum == int(guess):
                print(f"Congratulations you win! The number was: {chosenNum}")
                break
        
        loopGame = input("Play again? Y/N: ")
        if loopGame.upper() == "Y":
            self.Play()
        elif loopGame.upper() == "N":
            print("Thank you for playing")
        else:
            raise Exception("Input must be either 'Y' or 'N'")
```

```python
Kevin = Numbers()
Kevin.Play()
```
